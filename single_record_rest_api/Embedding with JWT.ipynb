### Embeddings calculation authenticated via JWT

```python
import requests
import os
import sys
from snowflake.snowpark import Session
import json
import datetime
from accessviakeypair import get_token
```

```python
account = "sfsenorthamerica-demo391"
user = "PLAKHANPAL"
private_key_file_path = "./rsa_key.p8"
endpoint = "nd64aqc-sfsenorthamerica-demo391.snowflakecomputing.app"
role = "SYSADMIN"
endpoint_path = "/encode"
spcs_url=f'https://{endpoint}{endpoint_path}'
data = {'data': [[0, 'This is the first sentence.'], [1, "Here's another sentence for testing."], [2, 'The quick brown fox jumps over the lazy dog.'], [3, 'I love coding and programming.'], [4, 'Machine learning is an exciting field.'], [5, 'Python is a popular programming language.'], [6, 'I enjoy working with data.'], [7, 'Deep learning models are powerful.'], [8, 'Natural language processing is fascinating.'], [9, 'I want to improve my NLP skills.']]}
print(spcs_url)
```

```python
token = get_token(account, user, private_key_file_path, endpoint, role)
print(token)
```

```python
def embed(token, url, data):
  # Create a request to the ingress endpoint with authz.
  headers = {'Authorization': f'Snowflake Token="{token}"'}
  response = requests.post(f'{url}',  json=data, headers=headers)
  return response.text
```

```python
%%time
results = embed(token=token, url=spcs_url, data=data)
print(results)
```
